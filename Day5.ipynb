```python
import re
from copy import deepcopy
```

```python
input = open('/content/drive/MyDrive/ERNI/AdventOfCode/data/input5.txt',"r")
```

```python
lines = []
line = 'X'
while(line != '\n'):
  line = input.readline()
  lines.append(line)
```

```python
lines = lines[0:len(lines)-1]
```

```python
stacks = {}
stack_to_go = 1
for l in range(0, len(lines)*4, 4):
  for x in range(9):
    if lines[x][l+1].isdigit():
        stack_to_go = int(lines[x][l+1])
    elif lines[x][l+1] not in [' ', '[', ']', '\n']:
      if stack_to_go in stacks.keys():
        stacks[stack_to_go].append(lines[x][l+1])
      else:
        stacks[stack_to_go] = [lines[x][l+1]]
          
        
  stack_to_go += 1
print(stacks)
```

```python
stacks1 = deepcopy(stacks)
stacks2 = deepcopy(stacks)
```

```python
move_lines = input.readlines()
```

```python
for m in range(len(move_lines)):
  move_subs = move_lines[m].split()
  piece_idx = int(move_subs[1])
  stack_idx = int(move_subs[3])
  dest_idx = int(move_subs[5])
  elements = stacks1[stack_idx][0:piece_idx]
  elements2 = stacks2[stack_idx][0:piece_idx]
  del stacks1[stack_idx][0:piece_idx]
  del stacks2[stack_idx][0:piece_idx]
  stacks1[dest_idx] = elements[::-1] + stacks1[dest_idx]
  stacks2[dest_idx] = elements2 + stacks2[dest_idx]
```
