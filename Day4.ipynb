```python
input = open('/content/drive/MyDrive/ERNI/AdventOfCode/data/input4.txt',"r")
```

```python
lines = input.readlines()
total = 0
total2 = 0
for i in range(0, len(lines)):

  line = lines[i].strip().split(',')
  s1, s2 = line[0].strip().split('-')
  f1, f2 = line[1].strip().split('-')
  s1,s2,f1,f2 = [int(x) for x in [s1,s2,f1,f2]]

  if s1<=f1 and f2<=s2 or f1<=s1 and s2<=f2:
    total += 1

  #part 2
  if s2 >= f1 and f2 >= s1:
    total2 += 1


print(total)
print(total2)
```
