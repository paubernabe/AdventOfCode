```python
from collections import Counter
```

```python
input = open('/content/drive/MyDrive/ERNI/AdventOfCode/data/input6.txt',"r")
```

```python
lines = input.readlines()
lines
```

```python

def message_packet(window):
  seq = ''
  acc = 0
  for l in range(0, len(lines[0])):
    if seq == '':
      seq = lines[0][l:l+window]
    else:
      seq = seq + lines[0][l]
    acc += 1
    freq = Counter(seq)
    if (len(seq) == len(freq)):
      print(seq)
      return acc
    else:
      seq = seq[1:l+window]




  
```

```python
print(message_packet(window = 4))
print(message_packet(window = 14))
```
