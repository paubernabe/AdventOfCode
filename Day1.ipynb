```python
input = open('/content/drive/MyDrive/ERNI/AdventOfCode/data/input.txt',"r")
```

```python
#PART 1
max_calories = 0
acc = 0
for l in input:
  if l in ['\n', '\r\n', '']:
    if (acc > max_calories):
      max_calories = acc
    acc = 0
  else:
    acc += int(l.rstrip())
print(max_calories)
```

```python
#PART 2
import numpy as np
calories = []
acc = 0
for l in input:
  if l in ['\n', '\r\n', '']:
    calories.append(acc)
    acc = 0
  else:
    acc += int(l.rstrip())
calories.sort(reverse=True)

print(calories[0:3])
sum(calories[0:3])
```
