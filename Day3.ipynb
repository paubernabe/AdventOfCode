```python
input = open('/content/drive/MyDrive/ERNI/AdventOfCode/data/input3.txt',"r")
```

```python
# PART 1
priorities = {}
ord_lower = 1
ord_upper = 27
for l in input:

  idx = int(((len(l)) -1) /2)
  subs0 = l[0:idx]
  subs1 = l[idx:len(l)-1]
  match = ''.join(set(subs0).intersection(subs1))
  
  if match.isupper():
    p = (ord(match) - ord('A')) + ord_upper
    if match in priorities.keys():
      priorities[match] = priorities[match] + p
    else:
      priorities[match] = p

  elif match.islower():
    p = (ord(match) - ord('a')) + ord_lower
    if match in priorities.keys():
      priorities[match] = priorities[match] + p
    else:
      priorities[match] = p


print(sum(priorities.values()))
```

```python
#PART 2

priorities = {}
ord_lower = 1
ord_upper = 27
lines = input.readlines()

for l in range(0, len(lines), 3):

  e0 = set(lines[l].strip())
  e1 = set(lines[l+1].strip())
  e2 = set(lines[l+2].strip())
  
  
  match = list(e1.intersection(e0, e2))[0]
  
  if match.isupper():
    p = (ord(match) - ord('A')) + ord_upper
    if match in priorities.keys():
      priorities[match] = priorities[match] + p
    else:
      priorities[match] = p

  elif match.islower():
    p = (ord(match) - ord('a')) + ord_lower
    if match in priorities.keys():
      priorities[match] = priorities[match] + p
    else:
      priorities[match] = p


print(sum(priorities.values()))
```
