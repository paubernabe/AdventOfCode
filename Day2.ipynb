```python
input = open('/content/drive/MyDrive/ERNI/AdventOfCode/data/input2.txt',"r")
```

```python
move_points = {'Y': 2, 'X': 1, 'Z': 3}
losing_moves = {'A': 'Z', 'B': 'X', 'C':'Y'} #r vs s, p vs r, p vs s
equivalence_dict = {'A': 'X', 'B': 'Y', 'C':'Z'}
total_score = 0
```

```python
#PART 1 15337

for l in input:
  oponent = l[0]
  you = l[2]
  if losing_moves[oponent] == you:
    total_score += move_points[you] #lose
  elif equivalence_dict[oponent] == you:
    total_score += move_points[you] + 3 #draw
  else:
    total_score += move_points[you] + 6 #win
print(total_score)
  

  
```

```python
#PART 2

move_points = {'B': 2, 'A': 1, 'C': 3}
game_points = {'Y': 3, 'X': 0, 'Z': 6}
moves = ['A', 'B', 'C']
total_score = 0

for l in input:
  oponent = l[0]
  result = l[2]
  if result == 'Y':
    total_score += (move_points[oponent] + game_points[result])
  elif result == 'X':
    total_score += (move_points[moves[(move_points[oponent]%len(moves)-1) - 1]] + game_points[result])
  elif result == 'Z':
    total_score += (move_points[moves[(move_points[oponent]%len(moves)-1) - 2]] + game_points[result])
print(total_score)
```
